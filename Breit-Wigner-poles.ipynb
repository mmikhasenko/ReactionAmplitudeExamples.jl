```julia
using Plots
```

```julia
function constructBW(mξ, mi, gi,ϕ = 0.0)

    function BW(s)
        mπ = 0.5
        1.0/(mi^2-s-1.0im*gi^2*
        sqrt(cis(ϕ)*(s-(mξ+mπ)^2))*sqrt(s-(mξ-mπ)^2)*cis(-ϕ/2.0)/
            (8π*s))
    end

    return BW
end

function plot_poles(amp)
    let sxv = linspace(1e-3, 2.0, 100), syv = linspace(-1.0, 0.3, 100)
        cal = [amp(sx+1.0im*sy) for sy in syv, sx in sxv]
        contour(sxv,syv, log.(abs.(cal)), levels=30)
    end
end
```

```julia
using Interact
```

```julia
@manipulate for g in linspace(1,6,30), mξ in linspace(0.15,0.9,10), m in linspace(0.5,1.5,30), ϕ in linspace(-π,π,100)
    plot_poles(constructBW(mξ,m,g,ϕ))
end
```
