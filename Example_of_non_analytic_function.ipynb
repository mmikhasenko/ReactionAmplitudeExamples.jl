```julia
f(z) = sqrt(1im*z-1)
g(a) = quadgk(f,-1,a)[1]
```

```julia
using Plots
using QuadGK
```

```julia
pyplot()
```

```julia
pl1 = let 
    plot([0,0],[-1,2], l=(3,:red),lab="",xlim=(-2,2),ylim=(-2,2))
    scatter!([0],[-1], m=(6,:red),lab="", ann=(-1,0.2,"\$-1\$"))
    ## integration contour
    plot!([-1+0im,0.5+0im], lab="",xlab="Re(z)",ylab="Im(z)", l=(2,:black), arrow=true, ann=(0.5,0.2,"\$a\$"))
    scatter!([-1],[0],m=(:c,:black), lab="")
    annotate!([(1.2,1.5,"\$f(a) = \\int_{-1}^{a}\\sqrt{iz-1}\\,\\mathrm{d} z\$")])
end
pl2 = let a=linspace(-1,2,100)
    cal = g.(a)
    plot(a, [real.(cal) imag.(cal)], lab=["Re f(a)" "Im f(a)"], xlab="a")
end
plot(pl1,pl2, layout=grid(1,2), size=(900,400))
```

```julia
savefig("nonanalytic_function.png")
```
