## Breit-Wigner amplitude with energy dependent width
### Stable particles

```julia
using Plots
```

The function is
$$
\frac{1}{m^2-s-im\Gamma(s)},\qquad \Gamma(s) = g^2\sqrt{(s-(m_1+m_2)^2)(s-(m_1+m_2)^2)}{16\pi s}
$$

```julia
function constructBW(mξ, mi, gi,ϕ = 0.0)
    function BW(s)
        mπ = 0.14
        1.0/(mi^2-s-1.0im*gi^2*
            sqrt(cis(ϕ)*(s-(mξ+mπ)^2))*sqrt(s-(mξ-mπ)^2)*cis(-ϕ/2.0) / (16π*s))
    end
    return BW
end

function plot_poles(amp)
    let sxv = linspace(-0.1, 3.0, 100), syv = linspace(-1.0, 0.3, 100)
        cal = [amp(sx+1.0im*sy) for sy in syv, sx in sxv]
        contour(sxv,syv, log.(abs.(cal)), levels=30, size=(800,600), xlab="real part", ylab="imag part")
    end
end
```

```julia
using Interact
```

```julia
amp = constructBW(mρ,m,g,0)
```

```julia
@manipulate for g in linspace(1,6,30), m in linspace(0.5,1.5,30)
    mρ=0.77
    amp = constructBW(mρ,m,g,0)
    plot(x->abs2(amp(x+1e-3im)), 0.91^2, 3.0, size=(800,600))
end
```

```julia
@manipulate for g in linspace(1,6,30), m in linspace(0.5,1.5,30), ϕ in linspace(-π,π,100)
    mρ=0.77
    plot_poles(constructBW(mρ,m,g,ϕ))
end
```
