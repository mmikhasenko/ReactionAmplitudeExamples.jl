```julia
using Plots
```

```julia
function ρ(s)
    mπ = 0.139;
    1/(8*π)*sqrt(1-4*mπ^2/s)
end

function BW(s)
    m = 0.7755;
    mπ = 0.139;
    g = 2.9;
    1/((m^2-s)/g^2-1im*ρ(s)/2)
end

function deck(s)
    s*exp(-2*s)
end
```

```julia
using QuadGK
```

Unitarity for t: $$\Delta t = 2t^* \rho\, t$$
Unitarity for f: $$\Delta f = 2t^* \rho\, f$$

Inhomogenious Omnes solution: 
$$
f = b + \frac{t}{\pi}\int \frac{b(s') \rho(s')}{s'-s}\mathrm{d} s'
$$

```julia
function uniD(s)
    mπ = 0.139;
    1/(2*π)*quadgk(sp->deck(sp)*ρ(sp)/(sp-s-0.0001im),4mπ^2,100)[1]
end
```

```julia
function amp(s)
    deck(s)+BW(s)*uniD(s)
end
```

```julia
plot(e->deck(e^2),0.3, 1.3,ls=:dash, lab="Deck", xlab="Mpipi (GeV)", ylab="amplitude")
plot!(e->real(amp(e^2)),0.3, 1.3, lab="Re@uniDeck")
plot!(e->imag(amp(e^2)),0.3, 1.3, lab="Im@uniDeck")
```

```julia
arg(f) = atan2(imag(f),real(f))
plot(e->180/π*arg(amp(e^2)),0.3, 1.3, lab="Phi@uniDeck")
```

```julia
plot(xlab="Mpipi (GeV)", ylab="amplitude")
plot!(e->real(BW(e^2)),0.3, 1.3, lab="Re@T")
plot!(e->imag(BW(e^2)),0.3, 1.3, lab="Im@T")
```

```julia
plot(e->abs(100*amp(e^2))^2, 0.3, 1.3, lab="Re@uniDeck")
plot!(e->abs(BW(e^2))^2, 0.3, 1.3, lab="Re@uniDeck")
```
