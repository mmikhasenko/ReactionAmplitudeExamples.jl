```julia
λ(x,y,z)=x^2+y^2+z^2-2*x*y-2*y*z-2*z*x;
s12(s,s23,cost,m1sq,m2sq,m3sq) = m1sq+m2sq+2*(s23+m2sq-m3sq)*(s+m1sq-s23)/(4*s23)-2*sqrt(λ(s23,m2sq,m3sq)*λ(s,m1sq,s23))/(4*s23)*cost;
```

```julia
using Plots
plotly()
using LaTeXStrings
```

```julia
mπ=0.14
mϕ=1.02
mρ = 0.77
fp = s23->s12(mϕ^2,s23,+1,mπ^2,mπ^2,mπ^2)
fm = s23->s12(mϕ^2,s23,-1,mπ^2,mπ^2,mπ^2)
plot(
    xlabel="s23 (GeV)", 
    ylabel="s12 (GeV)",
    aspect_ratio=1,
    xlim = (0, 0.9),
    ylim = (0, 0.9)
)
xr = linspace(4*mπ^2+1e-10,(mϕ-mπ)^2,100);
# plot!(, xr, lab="cosT=+1")
# plot!(fm, xr, lab="cosT=-1")
plot!([fp, fm], xr,
    line=:black,
    fillrange=[fm, fp], fillalpha=0.3, c=:orange,
    lab="phase space")
hline!([mρ^2], lab="rho+",line=(2,:dash))
vline!([mρ^2], lab="rho-",line=(2,:dash))
plot!(s23->mϕ^2+3*mπ^2-mρ^2-s23,linspace(4*mπ^2+1e-10,0.4,100),line=(2,:dash), lab="rho")
plot!(legend=:top, bglegend=:white, fglegend=:black)
```

```julia
mπ=0.14
mϕ=1.02
mρ = 0.77
pcut = 0.15
cut = mϕ^2+mπ^2-2*sqrt(pcut^2+mπ^2)*mϕ
fp = s23->s12(mϕ^2,s23,+1,mπ^2,mπ^2,mπ^2)
function fm2(s23)
    s12v = s12(mϕ^2,s23,-1,mπ^2,mπ^2,mπ^2)
    s12v < cut && return s12v
    return cut
end
plot(
    xlabel="s23 (GeV)", 
    ylabel="s12 (GeV)",
    aspect_ratio=1,
    xlim = (0, 0.9),
    ylim = (0, 0.9)
)
xr = linspace(4*mπ^2+1e-10,cut,100);
# plot!(, xr, lab="cosT=+1")
# plot!(fm, xr, lab="cosT=-1")
plot!([fp, fm2], xr,
    line=:black,
    fillrange=[fm2, fp], fillalpha=0.3, c=:orange,
    lab="phase space")
hline!([mρ^2], lab="rho+",line=(2,:dash))
vline!([mρ^2], lab="rho-",line=(2,:dash))
plot!(s23->mϕ^2+3*mπ^2-mρ^2-s23,linspace(4*mπ^2+1e-10,0.4,100),line=(2,:dash), lab="rho")
plot!(legend=:top, bglegend=:white, fglegend=:black)
```
